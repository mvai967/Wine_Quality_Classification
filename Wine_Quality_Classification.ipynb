Box_Plot

```python
plt.figure(1 , figsize = (20 , 9))

n =  0
for x in df.columns:
    n += 1 
    plt.subplot(4 , 3 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.boxplot(x = df[x])
    plt.title('{}'.format(x))
    
plt.show()
```

Observation --> Almost all of the features containes outliers , then just fixed it

```python
df.columns
```

```python
'''treating outliers '''
treating_feature = ['fixed_acidity' , 'volatile_acidity' , 'citric_acid' , 'residual_sugar' , 'chlorides' , 'free_sulfur_dioxide', 'total_sulfur_dioxide' , 'density' , 'pH' , 'sulphates' , 'alcohol']
''''outliers treatment - replacing with upper whisker
'''
for x in treating_feature:
    b = plt.boxplot(df[x])
    values = [item.get_ydata()[1] for item in b['whiskers']]
    upper_whisker = values[1]
    df[x][df[x] > upper_whisker] = upper_whisker

'''
lower outliers treatment - replacing with lower whisker
'''
treating_feature_lower = ['fixed_acidity' , 'volatile_acidity' , 'citric_acid' , 'residual_sugar' , 'chlorides' , 'free_sulfur_dioxide', 'total_sulfur_dioxide' , 'density' , 'pH' , 'sulphates' , 'alcohol']

for x in treating_feature_lower:
    b = plt.boxplot(df[x])
    values = [item.get_ydata()[1] for item in b['whiskers']]
    lower_whisker = values[0]
    df[x][df[x] < lower_whisker] = lower_whisker
```

```python
'''
Boxplot after outlier treatment
'''
plt.figure(1 , figsize = (20 , 9))

n =  0
for x in df.columns:
    n += 1 
    plt.subplot(4 , 3 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.boxplot(x = df[x])
    plt.title('{}'.format(x))
    
plt.show()
```

Start Model Building

Splitting Data into Train and Test

```python
X=df.drop('quality' ,axis=1)
y=df['quality']
```

```python
print(X.shape)
print(y.shape)
```

#find out feature importance

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
#plot graph of feature importance for better visualization
feat_imp=pd.Series(model.feature_importances_,index=X.columns)
feat_imp.nlargest(5).plot(kind='barh' , color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
print('Shape of training feature:', X_train.shape)
print('Shape of testing feature:', X_test.shape)
print('Shape of training label:', y_train.shape)
print('Shape of training label:', y_test.shape)
```

standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Scale trainint data
X_train = scaler.fit_transform(X_train)

# Scale test data
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
```

Model Evaluation

Model accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

Accuracy of the model is 0.64

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier()
dec = dec.fit(X_train,y_train)
```

```python
y_pred = dec.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

Accuracy of the model is 0.57

#after using hyperparameter tuning we can increase the accuracy of the model

Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
```

```python
rfc = RandomForestClassifier(n_estimators=80, bootstrap=False, class_weight="balanced" ,random_state=569)
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
cm=confusion_matrix(y_pred,y_test)
```

```python
cm
```

```python
print("Logistic Regression Accuracy" , '0.64')
print("Decision Tree Accuracy" , '0.54')
print('Random Forest Accuracy ', '0.64')
```
